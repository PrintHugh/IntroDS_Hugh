```python
import pandas as pd
```

```python
def pollutantmean (directory, pollutant, id = range(1,333)):
    sum = 0
    count = 0
    for location in id:
        str_location = str(location)
        length = len(str_location)
        if(length == 1):
            str_location = '00' + str_location
        if(length == 2):
            str_location = '0' + str_location
        
        file_name = directory + '/' + str_location + '.csv'
        print(file_name)
        df = pd.read_csv(file_name, sep = ',')             
        sum += df[pollutant].sum()
        count += df[pollutant].count()
        
        
    
    print(sum)
    print(count)
    average = sum/count
    print(average)
        
## Figure Out how to access the files which aren't 3 digit numbers because there are 0's infront of the number.
## Idea: Turn number into string and use len() to figure out how long
```

```python
average = pollutantmean('specdata','nitrate', range(23,24))
```

```python
def complete(directory, id = range(1,333)):
    final_df = []
    for location in id:
        str_location = str(location)
        length = len(str_location)
        if(length == 1):
            str_location = '00' + str_location
        if(length == 2):
            str_location = '0' + str_location
        
        file_name = directory + '/' + str_location + '.csv'
        df = pd.read_csv(file_name, sep = ',')   
        df = df.dropna()   
        final_df.append([location, len(df)])
        return final_df
```

```python

complete('specdata',range(3,4))
```

```python
def corr(directory, threshold):
    return_series = []
    for location in range(1,333):
        str_location = str(location)
        length = len(str_location)
        if(length == 1):
            str_location = '00' + str_location
        if(length == 2):
            str_location = '0' + str_location
        
        file_name = directory + '/' + str_location + '.csv'
        df = complete(directory, range(location, location+1))
        if(df[0][1] > threshold):
            file = pd.read_csv(file_name, sep = ',')
            sulfate = file.loc[: ,'sulfate']
            nitrate = file.loc[: ,'nitrate']
            return_series.append(sulfate.corr(nitrate))
    
    return return_series
        
            
    
```

```python
print(corr('specdata', 150))
```
