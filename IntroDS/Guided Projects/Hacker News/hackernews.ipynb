# Hacker News
- Hacker News is a site started by the startup incubator Y Combinator, where user-submitted stories (known as "posts") are voted and commented upon, similar to reddit.

## Opening and Reading the Data
- Here we open the dataset and turn it into a list which we then put in the variable hn.



```python
from csv import reader

open_file = open('untitled.txt', encoding='utf-8')
#read_file = reader(open_file)
hn = open_file.read().splitlines()
print(hn[:5])
```

## Removing the Header
- In these four lines of code we get rid of the header from the dataset because it doesn't include any information we need. Then we print out the header and the rest of the data

```python
header = hn[0]
hn = hn[1:]
print(header)
print(hn[:5])
```

## Seperating the Data
- Here we seperate the data base on their titles. If it starts with 'Ask Hn' we put it in the ask_posts list, titles that start with 'Show Hn' will go in the show_posts list and other titles will go in the other_posts list. To make sure capitilzation doesnt mess with the data we turn every letter to lowercase.

```python
ask_posts = []
show_posts = []
other_posts = []
for row in hn:
    row = row.split('\t')
    title = row[1]
    title = title.lower()
    if title.startswith('ask hn'):
        ask_posts.append(row)
    elif title.startswith('show hn'):
        show_posts.append(row)
    else:
        other_posts.append(row)
print("Ask Posts: ", len(ask_posts)) 
print("Show Posts: ", len(show_posts)) 
print("Other Posts: ", len(other_posts)) 
```

```python
colonCounter = 0
for post in ask_posts:
    title = post[1]
    title_list = title.split()
    title_list = title_list[2:]
    title_str = ''
    for word in title_list:
        title_str = title_str + word
    if('how' in title_str):
        colonCounter += 1

print(colonCounter)        
```

## Finding The Average Amount Of Comments for each type of Post
- Here we find the average number of posts for each type of post by dividing the total number of posts by the total number of comments.

```python
total_ask_comments = 0
for post in ask_posts:
    total_ask_comments += int(post[4])
avg_ask_comments = total_ask_comments/len(ask_posts)
print(avg_ask_comments)

total_show_comments = 0
for post in show_posts:
    total_show_comments += int(post[4])
avg_show_comments = total_show_comments/len(show_posts)
print(avg_show_comments)
```

```python
import datetime as dt
result_list = []
for post in ask_posts:
    thing_to_append = []
    thing_to_append.append(post[6])
    thing_to_append.append(int(post[4]))
    result_list.append(thing_to_append)
    
counts_by_hour = {}
comments_by_hour = {}

for row in result_list:
    time = dt.datetime.strptime(row[0], "%m/%d/%Y %H:%M")
    hour = dt.datetime.strftime(time, '%H')
    if hour not in counts_by_hour:
        counts_by_hour[hour] = 1
        comments_by_hour[hour] = row[1]
    else:
        counts_by_hour[hour] += 1
        comments_by_hour[hour] += row[1]
    
```

```python
avg_by_hour = []
for hour in counts_by_hour:
    avg_by_hour.append([hour, comments_by_hour[hour]/counts_by_hour[hour]])
print (avg_by_hour)
```

```python
swap_avg_by_hour = []
for hour in avg_by_hour:
    swap_avg_by_hour.append([hour[1], hour[0]])
sorted_swap = sorted(swap_avg_by_hour, reverse = True)
print("Top 5 Hours for Ask Posts Comments")
sorted_swap = sorted_swap[:5]
for row in sorted_swap:
    hour = dt.datetime.strptime(row[1], '%H')
    hour = dt.datetime.strftime(hour, '%H')
    print('{0}:00: {1} average comments per post.'.format(hour,"%.2f" % row[0]))
```

```python
import string
punctuation = string.punctuation
punctuation_list = []
for c in punctuation:
    punctuation_list.append(c)
    
bad_words = ['but','though','and','nevertheless','i','to','for','a','an','therfore','you','what','do','is','how','who','the'
             ,'your','of','are','at','in','why','be','have','']
```

```python
def removeBadWords(keywordList):
    newKeywordList = []
    repeatedWords = []
    
    for keyword in keywordList:
        keyword = keyword.lower()
        for punc in punctuation_list:
            if punc in keyword:
                keyword = keyword.replace(punc,'')
            for word in bad_words:                
                if keyword == word:
                    keyword = ''
        if(keyword not in repeatedWords):
            if(keyword != ''):
                newKeywordList.append(keyword)
        repeatedWords.append(keyword)
        
    return newKeywordList
                
        
```

```python
def CharPosCounter(character, list):
    character = character.lower()
    counter = 0
    combined_list = ''
    for word in list:
        combined_list = combined_list + word
    for letter in combined_list:
        if(character == letter):
            counter = counter+1
    return counter
    
        
```

```python
def PhrasePosCounter(phrase, list):
    phrase = phrase.lower()
    phrase_list = phrase.split()
    phrase_combined = ''
    combined_list = ''
    first_word_index = int()
    positions_list = []
    for i in range(0,len(list)):
        list[i] = list[i].lower()
    for phrase_word in phrase_list:
        phrase_combined = phrase_combined + phrase_word
    for list_word in list:
        combined_list = combined_list + list_word
        
    if phrase_combined in combined_list:
        if(phrase in list):
            first_word_index = list.index(phrase_list[0])
            for i in range(0,len(phrase_list)):
                positions_list.append(first_word_index + i)
            return positions_list
        else:
            return[-1]
    else:
        return [-1]
```
